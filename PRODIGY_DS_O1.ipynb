```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (adjust the path to your actual downloaded file)
file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2.csv'  # Replace with your actual file name
data = pd.read_csv("/content/drive/MyDrive/API_SP.POP.TOTL_DS2_en_csv_v2_85220.csv", skiprows=4)

# Extract relevant columns (latest population, e.g., 2022)
population_2022 = data[['Country Name', '2022']].dropna()

# Sort and select top 10 countries by population
top_10 = population_2022.sort_values(by='2022', ascending=False).head(10)

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(top_10['Country Name'], top_10['2022'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Most Populous Countries (2022)')
plt.xlabel('Country')
plt.ylabel('Population')
plt.tight_layout()
plt.show()
```
